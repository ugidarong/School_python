```python
import matplotlib.pyplot as plt
plt.show()
```

```python
#인라인 표시로 그리기
fig = plt.figure()
ax = fig.add_subplot(111)

dat = [0, 1]
ax.plot(dat)

#그림을 인라인 표시
plt.show()
```

```python
#add_subplot으로 서브 플롯 배치
#--------------------------------
#피겨 생성
fig = plt.figure()

#피겨 안에 서브플롯을 3개 배치한다.
ax1 = fig.add_subplot(221) #2행 2열 1번
ax2 = fig.add_subplot(222) #2행 2열 2번
ax3 = fig.add_subplot(223) #2행 2열 3번

plt.show()
```

```python
#서브 플롯 번호 할당 순서의 확인
#ㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡㅡ
fig = plt.figure()

#서브플롯 작성
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)

#번호 기술
for i, ax in enumerate([ax1, ax2, ax3], start = 1):
    txt = 'ax{0}\n(22{0})'.format(i)
    ax.text(0.2, 0.4, txt, fontsize = 24)

plt.show()
```

```python
#add_subplot()메서드에 따른 서브플롯의 배치(콤마 단락 기법)
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
```

```python
#subplots()에 따른 서브플롯 배치
#----------------------------------------------
#피겨 오브젝트 작성과 서브플롯 배치를 동시에 실행
fig, axes = plt.subplots(2, 2)
print(type(axes),axes)
plt.show()
```

```python
#행렬로 서브플롯에 서브플롯 타이틀을 지정
#----------------------------------------
fig,axes = plt.subplots(2, 2)
axes[0, 1].set_title('Subplot 0-1')
plt.show()
```

```python
#ggplot 스타일로 그리기
#------------------------------------
#스타일 적용
plt.style.use('ggplot')
fig = plt.figure()
ax = fig.add_subplot(111)

dat= [0,1]
ax.plot(dat)

plt.show()
```
