```python
import os 
import pandas as pd
base_url = 'C:/Users/scsoo/OneDrive/바탕 화면/sample-data-master/sample-data-master/anime'
anime_csv = os.path.join(base_url, 'anime.csv')
df = pd.read_csv(anime_csv)

df.loc[df['episodes'] == 'Unknown'].head()
```

```python
#DataFrmae.where()의 사용법
#해당하지 않는 데이터를 NaN으로 채운 dataframe을 리턴
df.where(df['rating']<9.2).head()
```

```python
#행, 열을 지정해서 값 바꾸기
import numpy as np
df.loc[74, 'episodes'] = np.nan
df.loc[74, 'episodes']
```

```python
#복수의 값 변경
df.loc[df['episodes'] == 'Unknown','episodes'] = np.nan
df.loc[df['episodes'] == 'Unknown','episodes']
```

```python
#결손값 제외하기
df.loc[df['episodes'].isnull()]
```

```python
#결손값 데이터 제외
df.dropna().loc[70:].head()
```

```python
#dropna는 비파괴적 조작이다.
#옵션으로 inplace를 집어넣으면 파괴적 조합이 된다
df.dropna(inplace=True)
df.loc[70:].head()
#파괴적 조작을 하면 참조되어 있는 모든 것이 수정되기에 주의가 필요
```
