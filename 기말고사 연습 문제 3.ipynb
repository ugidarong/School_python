```python
#아래 코드를 완성하여 ANimal 클래스를 상속받는 Dog 클래스를 정의하세요
class Animal:
    def __init__(self,name):
        self.name = name

    def printName(self):
        print(self.name)

class Dog(Animal):
    def __init__(self,name,sound):
        super().__init__(name)
        self.sound = sound

    def printSound(self):
        print(self.sound)

dog = Dog("행복이", "멍멍")
dog.printName()
dog.printSound()
```

```python
import math
import random

def sphere_volume_and_surface(*args):
    radius = random.choice(args)
    volume = (4/3) * math.pi * math.pow(radius,3)
    surface_area = 4 * math.pi * math.pow(radius,2)
    return radius, volume, surface_area

radius, vol, surf = sphere_volume_and_surface(4,5,6,99,123,323,439)
print("선택된 반지름 : %d -> 구의 부피 : %.2f , 겉넓이 : %.2f"%(radius, vol, surf))
```

```python
#아래 코드를 완성하여 상속 관계에서 슈퍼를 활요해 ㅜ모 클래스의 속성과 메소드를
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        return f"이름 : {self.name}, 나이 : {self.age}"

class Employee(Person):
    def __init__(self, name, age, department):
        super().__init__(name,age)
        self.department = department

    def introduce(self):
        base_intro = Person.introduce(self)
        return f"{base_intro}, 부서 : {self.department}"

class Manager(Employee):
    def __init__(self, name, age, department, level):
        super().__init__(name, age, department)
        self.level = level

    def introduce(self):
        base_intro = Employee.introduce(self)
        return f"{base_intro}, 관리 등급 : {self.level}"

manager = Manager("김철수", 40, "HR", "1급")
print(manager.introduce())
```

```python
class Parent:
    def __init__(self,name):
        self.name = name

    def greet(self):
        print("Hello, my name is", self.name)

class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)
        self.age = age

    def showInfo(self):
        Parent.greet(self)
        print("I am", self.age, "years old.")

c = Child("홍길동", 15)
c.showInfo()
```
