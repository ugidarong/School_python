```python
#아래 코드를 완성하고 Vehicle과 그 하위 클래스에서 메소드를 확장하고 새로운 동작을 .....
class Vehicle:
    def __init__(self,brand,model):
        self.brand = brand
        self.model = model

    def describe(self):
        return f"브랜드 : {self.brand}, 모델 : {self.model}"

class Car(Vehicle):
    def __init__(self,brand,model,max_speed):
        super().__init__(brand,model)
        self.max_speed = max_speed

    def describe(self):
        base_desc = Vehicle.describe(self)
        return f"{base_desc}, 최고속도 : {self.max_speed}Km/h"

    def start(self):
        return f"{self.model}가 출발합니다"

class ElectricCar(Car):
    def __init__(self,brand,model,max_speed,battery_capacity):
        super().__init__(brand,model,max_speed)
        self.battery_capacity = battery_capacity

    def describe(self):
        base_desc = Car.describe(self)
        return f"{base_desc}, 배터리 용량 : {self.battery_capacity}Kwh"

    def charge(self):
        return f"{self.model}의 배터리를 충전합니다."

tesla = ElectricCar("Tesla", "Model 3", 250, 75)
print(tesla.describe())
print(tesla.start())
print(tesla.charge())
```

```python
#아래 코드를  완성하여 car 클래스를 정의하고, 객체를 생성해 속성을 출력하시오
class Car:
    def __init__(self,brand,model):
        self.b = brand
        self.m = model

my_car = Car("Tesla", "Model 3")
print("브랜드",my_car.b)
print("모델",my_car.m)
```

```python
#예상 출력결과가 나올 수 있도록 빈칸을 채우세요
def calculate_average(*args):
    total = sum(args)
    count = len(args)
    return total/count

print(calculate_average(10,20,30))
print(calculate_average(10,30))
```
